# **Types of Bar Chart**

# 1. Vertical Bar Chart
 - Displays categories vertically.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Plot
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart')
plt.show()
```

# 2. Horizontal Bar Chart
 - Useful for larger labels.

```python
# Plot
plt.barh(categories, values, color='orange')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
```

# 3. Stacked Bar Chart
 - Shows data composition across categories.

```python
# Data
values_1 = [5, 7, 3, 8]
values_2 = [3, 4, 2, 6]

# Plot
plt.bar(categories, values_1, color='skyblue', label='Set 1')
plt.bar(categories, values_2, bottom=values_1, color='lightgreen', label='Set 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

# 4. Grouped Bar Chart
 - Compares multiple datasets side by side.

```python
import numpy as np

# Data
x = np.arange(len(categories))  # x locations for the groups
width = 0.35                    # width of the bars

# Plot
plt.bar(x - width/2, values_1, width, label='Set 1', color='skyblue')
plt.bar(x + width/2, values_2, width, label='Set 2', color='lightgreen')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

# 5. 100% Stacked Bar Chart
 - Highlights proportional data distribution.

```python
# Data (normalized to percentages)
total = [i + j for i, j in zip(values_1, values_2)]
percentages_1 = [i / j * 100 for i, j in zip(values_1, total)]
percentages_2 = [i / j * 100 for i, j in zip(values_2, total)]

# Plot
plt.bar(categories, percentages_1, color='skyblue', label='Set 1')
plt.bar(categories, percentages_2, bottom=percentages_1, color='lightgreen', label='Set 2')
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('100% Stacked Bar Chart')
plt.legend()
plt.show()
```

# 6. Lollipop Chart
 - Simplifies value display with markers and lines.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Plot
plt.stem(categories, values, basefmt=" ", markerfmt="o", linefmt='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart')
plt.show()
```

# 7. Divergent Bar Chart
 - Ideal for positive and negative value distinctions.

```python
# Divergent Data
divergent_values = [5, -3, 7, -4]

# Plot
plt.bar(categories, divergent_values, color=['skyblue' if v > 0 else 'salmon' for v in divergent_values])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Divergent Bar Chart')
plt.axhline(0, color='black', linewidth=0.5)
plt.show()
```

# 8. Gantt Chart
 - Used for project timelines.

```python
import matplotlib.dates as mdates
import datetime

# Sample Data
tasks = ['Task A', 'Task B', 'Task C']
start_dates = [datetime.date(2023, 5, 1), datetime.date(2023, 5, 10), datetime.date(2023, 5, 15)]
durations = [10, 5, 7]

# Plot
fig, ax = plt.subplots()
for i, task in enumerate(tasks):
    ax.barh(task, durations[i], left=start_dates[i], color='skyblue')

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
plt.xlabel('Date')
plt.ylabel('Tasks')
plt.title('Gantt Chart')
plt.show()
```

# 9. Floating Bar Chart
 - Shows ranges within each category

```python
# Data with start and end points
start = [2, 5, 7, 3]
end = [5, 8, 10, 6]
diff = [end[i] - start[i] for i in range(len(start))]

# Plot
plt.bar(categories, diff, bottom=start, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Range')
plt.title('Floating Bar Chart')
plt.show()
```

# Practical - 2
 - With Iris DataSet

```python
# Import Required Libraries
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.dates as mdates
```

```python
# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df['species'] = iris.target
```

```python
# Extracting sample data for plotting
categories = iris_df['species'].value_counts().index.map(lambda x: iris.target_names[x])
values = iris_df['species'].value_counts().values
values_1 = values
values_2 = np.array([10, 15, 20])  # Sample secondary values for stacked/grouped bar charts
x = np.arange(len(categories))
width = 0.35

# Create a figure to contain all the plots
fig, axs = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('Examples of Different Types of Bar Charts with Iris Dataset', fontsize=16)

# 1. Vertical Bar Chart
axs[0, 0].bar(categories, values, color='skyblue')
axs[0, 0].set_title('Vertical Bar Chart')
axs[0, 0].set_xlabel('Species')
axs[0, 0].set_ylabel('Count')

# 2. Horizontal Bar Chart
axs[0, 1].barh(categories, values, color='orange')
axs[0, 1].set_title('Horizontal Bar Chart')
axs[0, 1].set_xlabel('Count')
axs[0, 1].set_ylabel('Species')

# 3. Stacked Bar Chart
axs[0, 2].bar(categories, values_1, color='skyblue', label='Set 1')
axs[0, 2].bar(categories, values_2, bottom=values_1, color='lightgreen', label='Set 2')
axs[0, 2].set_title('Stacked Bar Chart')
axs[0, 2].legend()

# 4. Grouped Bar Chart
axs[1, 0].bar(x - width/2, values_1, width, label='Set 1', color='skyblue')
axs[1, 0].bar(x + width/2, values_2, width, label='Set 2', color='lightgreen')
axs[1, 0].set_title('Grouped Bar Chart')
axs[1, 0].set_xticks(x)
axs[1, 0].set_xticklabels(categories)
axs[1, 0].legend()

# 5. 100% Stacked Bar Chart
total = [i + j for i, j in zip(values_1, values_2)]
percentages_1 = [i / j * 100 for i, j in zip(values_1, total)]
percentages_2 = [i / j * 100 for i, j in zip(values_2, total)]
axs[1, 1].bar(categories, percentages_1, color='skyblue', label='Set 1')
axs[1, 1].bar(categories, percentages_2, bottom=percentages_1, color='lightgreen', label='Set 2')
axs[1, 1].set_title('100% Stacked Bar Chart')
axs[1, 1].set_ylabel('Percentage')
axs[1, 1].legend()

# 6. Lollipop Chart
axs[1, 2].stem(categories, values, basefmt=" ", markerfmt="o", linefmt='skyblue')
axs[1, 2].set_title('Lollipop Chart')
axs[1, 2].set_xlabel('Species')
axs[1, 2].set_ylabel('Count')

# 7. Divergent Bar Chart
divergent_values = [15, -7, 10]  # Divergent values for illustration
axs[2, 0].bar(categories, divergent_values, color=['skyblue' if v > 0 else 'salmon' for v in divergent_values])
axs[2, 0].axhline(0, color='black', linewidth=0.5)
axs[2, 0].set_title('Divergent Bar Chart')

# 8. Gantt Chart (using example dates for illustration)
tasks = ['Setosa', 'Versicolor', 'Virginica']
start_dates = [datetime.date(2023, 5, 1), datetime.date(2023, 5, 10), datetime.date(2023, 5, 15)]
durations = [10, 5, 7]
for i, task in enumerate(tasks):
    axs[2, 1].barh(task, durations[i], left=start_dates[i], color='skyblue')
axs[2, 1].xaxis_date()
axs[2, 1].xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
axs[2, 1].set_title('Gantt Chart')

# 9. Floating Bar Chart
start = [2, 5, 7]
end = [5, 8, 10]
diff = [end[i] - start[i] for i in range(len(start))]
axs[2, 2].bar(categories, diff, bottom=start, color='skyblue')
axs[2, 2].set_title('Floating Bar Chart')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
